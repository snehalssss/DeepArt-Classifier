## CLEAN RAW DATA FILE

```python
art_data = pd.read_csv('../raw_data/artists.csv')
```

```python
#drop id, years, nationality, bio, wikipedia
art_data.drop(labels='id', axis=1, inplace = True)

#rename 'Albrecht Dürer' to 'Albrecht Durer'
art_data['name'] = art_data['name'].replace(to_replace='Albrecht D.rer', value = 'Albrecht Durer', regex = True)
art_data['name'] = art_data['name'].str.replace(' ', '_')
```

```python
art_data['name']
```

```python
#create a new folder for the csv files
datapath = '..\data'
create_new_folder(datapath)
```

```python
#create data frame with artist name, genre, filename, filepath
path = r'..\raw_images'
#get image information
art_file = get_image_info(path)
```

```python
art_file
```

```python
#extract the artist name and genre
labels = art_data.loc[:,['name', 'genre']]
labels['label'] = labels['name']
labels.drop(labels='name', axis=1, inplace=True)
```

```python
labels
```

```python
#label each file with their respective genres
art_file = art_file.merge(labels, how='left', on='label')
art_file.head()
```

```python
#drop artworks with multiple styles
art_file = art_file.loc[~art_file.genre.str.contains(',')]
art_file.reset_index(inplace=True, drop=True)
art_file.shape
```

### SAVE CSV FILEPATHS

```python
#create a csv file from the new dataframe
newfile = 'art_labels.csv'
fullpath = os.path.join(datapath, newfile)
art_file.to_csv(fullpath)
```

## SPLIT DATA INTO TRAIN, VALIDATION, AND TEST SETS

```python
import pandas as pd

# Define folder paths
images = r'../raw_images'  # This never changes, only for reference
train = r'../train_images'
val = r'../val_images'
test = r'../test_images'

# Get image metadata
images_df = get_image_info(images)

# Define the genres to sample
genre_sample = ['Impressionism', 'Cubism', 'Expressionism', 'Pop Art', 'Byzantine Art', 'Abstract Expressionism']

# Select only images that match the genre_sample
genre_images = art_file[art_file['genre'].isin(genre_sample)]  # Should contain 2,306 images
print(f"Total images matching genre_sample: {len(genre_images)}")  # Debug

# ---- TEST SET (10%) ----
print("\nProcessing test dataset...")
test_file = genre_images.sample(frac=0.10, random_state=42)  # Select exactly 10% (231 images)
get_sample(test, test_file, genre_sample, 1)  # Move test images
test_df = get_image_info(test)
print(f"Test images selected: {len(test_df)}")  # Should be 231

# ---- TRAIN SET (81%) ----
print("\nProcessing training dataset...")
train_file = genre_images[~genre_images.filename.isin(test_df.filename)]  # Remove test images

# Select exactly 1868 images for training
train_sample = train_file.sample(n=1868, random_state=42)
get_sample(train, train_sample, genre_sample, 1)  # Move train images
train_df = get_image_info(train)
train_df.to_csv(r'..\data\train.csv', index=False)
print(f"Train images selected: {len(train_df)}")  # Should be 1868

# ---- VALIDATION SET (9%) ----
print("\nProcessing validation dataset...")
val_file = train_file[~train_file.filename.isin(train_sample.filename)]  # Remaining images after train selection

# Select exactly 207 images for validation
val_sample = val_file.sample(n=207, random_state=42)  # Ensure exactly 207 images
get_sample(val, val_sample, genre_sample, 1)  # Move validation images
val_df = get_image_info(val)
print(f"Validation images selected: {len(val_df)}")  # Should be 207

# ---- Final Debug ----
print("\nFinal dataset sizes:")
print(f"Total genre images: {len(genre_images)}")  # 2306
print(f"Test images: {len(test_df)} (Expected: 231)")
print(f"Train images: {len(train_df)} (Expected: 1868)")
print(f"Validation images: {len(val_df)} (Expected: 207)")
```
